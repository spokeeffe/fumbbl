```python
#import pandas as pd
import requests
import json
```

```python
fapi = "https://fumbbl.com/api/"
cplus_group_id = 14081
```

```python
def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
```

```python
def json_to_dict(x):
    'convert json strings to dictionaries'
    if x is not None :
        return json.loads(x)
    else :
        return x
```

```python
#get all teams from within a group
cplus_teams = requests.get(fapi + "group/teams/" + str(cplus_group_id))
```

```python
jprint(cplus_teams.json())
```

```python
#pull teamids from group into a dict
cplus_teams_dict = cplus_teams.json()
print(cplus_teams_dict['teams'][0]['team'])
```

```python
def json_to_list(response, key, field):
    # takes an api response, and a key, and returns all values for a given field in that key
    rdict = response.json()
    new_list=[]
    for record in rdict[key]:
        s=record[field]
        new_list.append(s)
    return new_list
```

```python
cplus_team_ids = json_to_list(cplus_teams, 'teams', 'team')
```

```python
print(cplus_team_ids)
```
