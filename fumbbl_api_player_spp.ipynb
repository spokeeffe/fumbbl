```python
###############
# 1 -PACKAGES #
###############

#import pandas as pd
import requests
import json
import csv

fapi = "https://fumbbl.com/api/"
cplus_group_id = 14081
cibbl_group_id = 10455
tournament_id = 55528

group_id = cibbl_group_id
```

```python
################
# 2 -FUNCTIONS #
################

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
    
    
def json_to_list(response, key, field):
    # takes an api response, and a key, and returns all values for a given field in that key
    rdict = response.json()
    new_list=[]
    for record in rdict[key]:
        s=record[field]
        new_list.append(s)
    return new_list


def print_player_spp(team_id_list, delimiter, csv_name):
    #iterate through list of teams and call player_team_spp for each
    for record in team_id_list:
        team_response = get_team_object(record)
        team_player_spp(team_response, delimiter, csv_name)

        
def team_player_spp(response, delimiter, csv_name):
    #given a team json response, pull player_name, team_name, spp and write to specified .csv
    rdict = response.json()
    team_name = rdict['name']
    
    #with open(csv_name, 'w', encoding='UTF8') as f:
    f = open(csv_name, 'a', encoding='UTF8')
    writer = csv.writer(f)
    
    for record in rdict['players']:
        row_list = []
        player_name = record['name']
        player_spp = record['record']['spp']
        row_list.append(team_name)
        row_list.append(player_name)
        row_list.append(player_spp)
        #data_string = team_name + delimiter + player_name + delimiter + str(player_spp)
        
        # write the data to csv
        writer.writerow(row_list)
    f.close()


def get_team_object(team_id):
    #given a team_id retrieve the "team" api response
    team_response = requests.get(fapi +"team/get/" + team_id)
    return team_response


def schedule_to_unique_team_ids(response):
    sdict = response.json()
    new_list=[]
    for i in sdict:
        t1=i["teams"][0]["id"]
        t2=i["teams"][1]["id"]
        new_list.append(str(t1))
        new_list.append(str(t2))
    
    teamlist = list(dict.fromkeys(new_list))
    return teamlist
```

```python
##############
# 2 -GET SPP #
##############

# get all teams from within a group
teams = requests.get(fapi + "group/teams/" + str(group_id))
# get a list of the team_ids from the group
team_ids = json_to_list(teams, 'teams', 'team')
# get team and player spp written to specified file
print_player_spp(team_ids, '|', 'cibbl_spp_list.csv')
```
