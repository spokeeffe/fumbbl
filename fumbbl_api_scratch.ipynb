```python
#import pandas as pd
import requests
import json
```

```python
def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)
```

```python
outcasts = requests.get("https://fumbbl.com/api/team/get/950292")
```

```python
jprint(outcasts.json())
```
