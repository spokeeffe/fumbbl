```python
def get_players_ids_from_team(team_id):
    api_url = f"https://fumbbl.com/api/team/get/{team_id}"
    print(api_url)
    player_id_list = []
    team = requests.get(api_url)
    for player in team.json()["players"]:
        player_id_list.append(player.get('id'))
        
    return player_id_list
```

```python
get_players_ids_from_team(950292)
```

```python
def get_match_ids_from_tournament(tournament_id):
    api_url = f"https://fumbbl.com/api/tournament/schedule/{tournament_id}"
    print(api_url)
    match_id_list = []
    schedule = requests.get(api_url)
    for match in schedule.json():
        match_id_list.append(match["result"].get('id'))
        
    return match_id_list
```
